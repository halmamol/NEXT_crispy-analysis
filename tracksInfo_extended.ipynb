```python
d_DS = np.load(npz_path+npz_filename_DS)
d_DS_tb = np.load(npz_path+npz_filename_DS_tb)
d_DS_tb2 = np.load(npz_path+npz_filename_DS_tb2)
nt_0nubb_DS = d_DS['ntracks_0nubb']
nt_0nubb_DS_tb = d_DS_tb['ntracks_0nubb']
nt_0nubb_DS_tb2 = d_DS_tb2['ntracks_0nubb']

nt_eroi_DS = d_DS['ntracks_eroi']
nt_eroi_DS_tb = d_DS_tb['ntracks_eroi']
nt_eroi_DS_tb2 = d_DS_tb2['ntracks_eroi']
```
