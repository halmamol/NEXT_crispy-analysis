```python
npz_path = '/Users/halmamol/NEXT/NBarrays/'
npz_filename_direct = f'neutrons_center.npz' 
npz_filename_4pi = f'neutrons_4pi.npz'
```

```python
d_direct = np.load(npz_path+npz_filename_direct)
d_4pi = np.load(npz_path+npz_filename_4pi)

direct_500keV = d_direct['detector_energy_500keV']
direct_2MeV = d_direct['detector_energy_2MeV']

d4pi_500keV = d_4pi['detector_energy_500keV']
d4pi_2MeV = d_4pi['detector_energy_2MeV']
```

```python
plt.hist(direct_500keV, bins = 100, range=(5,600), histtype='step', color = 'red', label = 'center - 500keV', linestyle='dashed')
plt.hist(direct_2MeV, bins = 100, range=(5,600), histtype='step', color = 'black', label = 'center - 2MeV')
plt.hist(d4pi_500keV, bins = 100, range=(5,600), alpha=0.8, color='gray', label = '4pi - 500keV')
plt.hist(d4pi_2MeV, bins = 100, range=(5,600), histtype='step', color='blue', label = '4pi - 2MeV')

plt.xlabel('events energy (keV)')
plt.ylabel('events (a.u.)')
#plt.title('Neutrons pointing to center detector')
#plt.yscale('log')
plt.legend(loc=1);
plt.show()
```
